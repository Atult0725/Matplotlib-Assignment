```python
'''Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data 
plot a scatter plot.

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot.
'''

import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x=3+np.random.normal(0,2,50)
y=3+np.random.normal(0,2,len(x))
plt.scatter(x,y)
plt.xlabel('This is x column')
plt.ylabel('This is y column')
plt.title("Scatter plot")
plt.show()

```

```python

'''Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

Use the following data:

import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and  y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])


Ans-The subplot() function is used to create multiple plots in the same figure. It allows us to compare multiple plots side-by-side or create a complex visualization with multiple subplots.

Here is an example of creating four line plots using the subplot() function with the given.

'''

import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

fig,axe=plt.subplots(2,2)
axe[0,0].plot(x1,y1)
axe[0,0].set_title("Line 1")

axe[0,1].plot(x2,y2)
axe[0,1].set_title("Line 2")

axe[1,0].plot(x3,y3)
axe[1,0].set_title("Line 3")

axe[1,1].plot(x4,y4)
axe[1,1].set_title('Line 4')

plt.subplots_adjust(wspace=0.3, hspace=0.3)

plt.show()

```

```python

'''Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot. 

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

ANs--A bar plot, also known as a bar chart, is a graphical representation of data that uses rectangular bars to represent the values of a categorical variable. The height or length of each bar represents the value of the variable being plotted. Bar plots are commonly used to display and compare data across different categories or groups.

Bar plots are used to:

Compare values across different categories or groups
Display data that can be easily represented as discrete categories
Show changes in data over time or across different conditions
Here's how you can create a bar plot and a horizontal bar plot using the given data:


'''

import numpy as np
import matplotlib.pyplot as plt

# Define the data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a bar plot
plt.bar(company, profit)
plt.title('Company Profit')
plt.xlabel('Company')
plt.ylabel('Profit (in millions)')
plt.show()


```

```python

'''Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

ANs-A box plot, also known as a box-and-whisker plot, is a graphical representation of data that displays the distribution of a continuous variable through its quartiles. The box in the plot represents the interquartile range (IQR) of the data, with the line in the middle of the box representing the median. The whiskers extending from the box represent the range of the data outside of the IQR, and any points beyond the whiskers are considered outliers.

Box plots are used to:

Visualize the distribution and spread of data
Compare the distributions of different groups or samples
Identify outliers and extreme values
Here's how you can create a box plot using the given data:


'''
import numpy as np
import matplotlib.pyplot as plt

# Define the data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Combine the data into a list
data = [box1, box2]

# Create a box plot
plt.boxplot(data)
plt.title('Box Plot')
plt.xlabel('Box')
plt.ylabel('Value')
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.show()
```
